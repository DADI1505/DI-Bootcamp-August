```python
ny_data = df[df['State'] == 'New York']
ca_data = df[df['State'] == 'California']

ny_sales = ny_data['Sales'].sum()
ca_sales = ca_data['Sales'].sum()

ny_profit = ny_data['Profit'].sum()
ca_profit = ca_data['Profit'].sum()

sales_diff = ca_sales - ny_sales
profit_diff = ca_profit - ny_profit

print(f"Ventes totales New York: {ny_sales:.2f}")
print(f"Ventes totales Californie: {ca_sales:.2f}")
print(f"DiffÃ©rence de ventes (Californie - New York): {sales_diff:.2f}\n")

print(f"BÃ©nÃ©fices totaux New York: {ny_profit:.2f}")
print(f"BÃ©nÃ©fices totaux Californie: {ca_profit:.2f}")
print(f"DiffÃ©rence de bÃ©nÃ©fices (Californie - New York): {profit_diff:.2f}")
```

```python
customer_profit = df.groupby('Customer Name')['Profit'].sum().sort_values(ascending=False)

total_customers = len(customer_profit)
top_20_percent_customers = int(total_customers * 0.20)

profit_top_customers = customer_profit.head(top_20_percent_customers).sum()
total_profit = customer_profit.sum()

percentage_profit_by_top_customers = (profit_top_customers / total_profit) * 100
percentage_of_customers = (top_20_percent_customers / total_customers) * 100

print(f"Nombre total de clients: {total_customers}")
print(f"Nombre de clients reprÃ©sentant les 20%% supÃ©rieurs: {top_20_percent_customers} ({percentage_of_customers:.2f}%%)")
print(f"BÃ©nÃ©fice total de tous les clients: {total_profit:.2f}")
print(f"BÃ©nÃ©fice des {top_20_percent_customers} meilleurs clients: {profit_top_customers:.2f}")
print(f"Les {percentage_of_customers:.2f}%% des clients les plus rentables gÃ©nÃ¨rent {percentage_profit_by_top_customers:.2f}%% du bÃ©nÃ©fice total.")

if percentage_profit_by_top_customers >= 80:
    print("\nOui, le principe de Pareto s'applique aux clients et aux bÃ©nÃ©fices : les 20% des clients les plus rentables contribuent Ã  plus de 80% du bÃ©nÃ©fice total.")
else:
    print("\nNon, le principe de Pareto (rÃ¨gle 80/20) ne s'applique pas strictement aux clients et aux bÃ©nÃ©fices dans cet ensemble de donnÃ©es, car les 20% des clients les plus rentables ne gÃ©nÃ¨rent pas 80% ou plus du bÃ©nÃ©fice total.")
```

```python
city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False)
print("Top 20 des villes par ventes :\n")
print(city_sales.head(20))

city_profit = df.groupby('City')['Profit'].sum().sort_values(ascending=False)
print("\nTop 20 des villes par bÃ©nÃ©fices :\n")
print(city_profit.head(20))

top_20_sales_cities = set(city_sales.head(20).index)
top_20_profit_cities = set(city_profit.head(20).index)

common_cities = top_20_sales_cities.intersection(top_20_profit_cities)
sales_only_cities = top_20_sales_cities.difference(top_20_profit_cities)
profit_only_cities = top_20_profit_cities.difference(top_20_sales_cities)

print("\n--- Analyse des diffÃ©rences de rentabilitÃ© entre les villes ---")

if common_cities:
    print("\nVilles prÃ©sentes dans les Top 20 des ventes ET des bÃ©nÃ©fices :")
    for city in common_cities:
        sales = city_sales.loc[city]
        profit = city_profit.loc[city]
        print(f"  - {city}: Ventes = {sales:.2f}, BÃ©nÃ©fices = {profit:.2f}")
else:
    print("\nAucune ville n'est prÃ©sente Ã  la fois dans le Top 20 des ventes et le Top 20 des bÃ©nÃ©fices.")

if sales_only_cities:
    print("\nVilles dans le Top 20 des ventes mais PAS dans le Top 20 des bÃ©nÃ©fices :")
    for city in sales_only_cities:
        sales = city_sales.loc[city]
        profit = city_profit.loc[city]
        print(f"  - {city}: Ventes = {sales:.2f}, BÃ©nÃ©fices = {profit:.2f}")
else:
    print("\nToutes les villes du Top 20 des ventes sont Ã©galement dans le Top 20 des bÃ©nÃ©fices (ou il n'y a pas de villes uniques).")

if profit_only_cities:
    print("\nVilles dans le Top 20 des bÃ©nÃ©fices mais PAS dans le Top 20 des ventes :")
    for city in profit_only_cities:
        sales = city_sales.loc[city]
        profit = city_profit.loc[city]
        print(f"  - {city}: Ventes = {sales:.2f}, BÃ©nÃ©fices = {profit:.2f}")
else:
    print("\nToutes les villes du Top 20 des bÃ©nÃ©fices sont Ã©galement dans le Top 20 des ventes (ou il n'y a pas de villes uniques).")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate total sales for each customer and sort them
customer_sales = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False)

# Calculate cumulative sales percentage
total_sales_customers = customer_sales.sum()
cum_sales_percent = (customer_sales.cumsum() / total_sales_customers) * 100

# Calculate cumulative customer percentage
total_customers = len(customer_sales)
cum_customer_percent = (np.arange(1, total_customers + 1) / total_customers) * 100

# Plot the Pareto curve
plt.figure(figsize=(10, 6))
plt.plot(cum_customer_percent, cum_sales_percent, marker='o', linestyle='-')
plt.title('Courbe Cumulative des Ventes par Clients (Principe de Pareto)')
plt.xlabel('Pourcentage CumulÃ© de Clients')
plt.ylabel('Pourcentage CumulÃ© des Ventes')
plt.grid(True)
plt.axvline(20, color='r', linestyle='--', label='20% des Clients')
plt.axhline(80, color='g', linestyle='--', label='80% des Ventes')
plt.legend()
plt.show()

# Check the 80/20 rule
top_20_percent_customers_count = int(total_customers * 0.20)
sales_by_top_20_customers = customer_sales.head(top_20_percent_customers_count).sum()
percentage_sales_by_top_20_customers = (sales_by_top_20_customers / total_sales_customers) * 100

print(f"\nNombre total de clients: {total_customers}")
print(f"Nombre de clients reprÃ©sentant les 20%% supÃ©rieurs: {top_20_percent_customers_count}")
print(f"Total des ventes de tous les clients: {total_sales_customers:.2f}")
print(f"Ventes des {top_20_percent_customers_count} meilleurs clients: {sales_by_top_20_customers:.2f}")
print(f"Les {percentage_of_customers:.2f}%% des clients les plus performants gÃ©nÃ¨rent {percentage_sales_by_top_20_customers:.2f}%% du total des ventes.")

if percentage_sales_by_top_20_customers >= 80:
    print("\nOui, le principe de Pareto s'applique aux clients et aux ventes : les 20% des clients les plus performants contribuent Ã  plus de 80% du total des ventes.")
else:
    print("\nNon, le principe de Pareto (rÃ¨gle 80/20) ne s'applique pas strictement aux clients et aux ventes dans cet ensemble de donnÃ©es, car les 20% des clients les plus performants ne gÃ©nÃ¨rent pas 80% ou plus du total des ventes.")
```
