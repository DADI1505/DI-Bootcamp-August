## Analyse de zone : États et Villes

### Quels États ont le plus de ventes ?

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


df = pd.read_csv('/content/sample_data/Sample - Superstore.csv',encoding='latin1')
sales_by_state = df.groupby('State')['Sales'].sum().sort_values(ascending=False).reset_index()
print("Top 10 États par ventes :")
display(sales_by_state.head(10))

plt.figure(figsize=(14, 7))
sns.barplot(x='Sales', y='State', data=sales_by_state.head(10), palette='viridis')
plt.title('Top 10 États par ventes', fontsize=16)
plt.xlabel('Ventes totales', fontsize=12)
plt.ylabel('État', fontsize=12)
plt.tight_layout()
plt.show()
```

### Quelle est la différence entre New York et la Californie en termes de ventes et de bénéfices ?

```python
ny_ca_data = df[df['State'].isin(['New York', 'California'])]
ny_ca_summary = ny_ca_data.groupby('State').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

print("Comparaison des ventes et bénéfices entre New York et la Californie :")
display(ny_ca_summary)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.barplot(x='State', y='Sales', data=ny_ca_summary, palette='pastel', ax=axes[0])
axes[0].set_title('Comparaison des ventes (New York vs Californie)', fontsize=14)
axes[0].set_xlabel('État', fontsize=12)
axes[0].set_ylabel('Ventes totales', fontsize=12)

sns.barplot(x='State', y='Profit', data=ny_ca_summary, palette='coolwarm', ax=axes[1])
axes[1].set_title('Comparaison des bénéfices (New York vs Californie)', fontsize=14)
axes[1].set_xlabel('État', fontsize=12)
axes[1].set_ylabel('Bénéfices totaux', fontsize=12)

plt.tight_layout()
plt.show()
```

### Qui est un client exceptionnel à New York ?

```python
ny_customers = df[df['State'] == 'New York']
exceptional_ny_customer = ny_customers.groupby('Customer Name').agg({'Sales': 'sum', 'Profit': 'sum'}).sort_values(by='Sales', ascending=False).head(5).reset_index()

print("Top 5 clients exceptionnels à New York (basé sur les ventes) :")
display(exceptional_ny_customer)

plt.figure(figsize=(12, 6))
sns.barplot(x='Sales', y='Customer Name', data=exceptional_ny_customer, palette='mako')
plt.title('Top 5 clients à New York (par ventes)', fontsize=16)
plt.xlabel('Ventes totales', fontsize=12)
plt.ylabel('Nom du client', fontsize=12)
plt.tight_layout()
plt.show()
```

### Y a-t-il des différences entre les États en termes de rentabilité ?

```python
profit_by_state = df.groupby('State')['Profit'].sum().sort_values(ascending=False).reset_index()

print("Top 10 États par profit :")
display(profit_by_state.head(10))

print("10 États avec les profits les plus bas (ou pertes) :")
display(profit_by_state.tail(10))

plt.figure(figsize=(16, 8))
sns.barplot(x='Profit', y='State', data=profit_by_state.head(10), palette='Greens_r')
plt.title('Top 10 États par profit', fontsize=16)
plt.xlabel('Profit total', fontsize=12)
plt.ylabel('État', fontsize=12)
plt.tight_layout()
plt.show()

plt.figure(figsize=(16, 8))
sns.barplot(x='Profit', y='State', data=profit_by_state.tail(10), palette='Reds')
plt.title('10 États avec les profits les plus bas (ou pertes)', fontsize=16)
plt.xlabel('Profit total', fontsize=12)
plt.ylabel('État', fontsize=12)
plt.tight_layout()
plt.show()
```

### Application du principe de Pareto

#### Peut-on appliquer le principe de Pareto aux clients et au profit ?

```python
customer_profit = df.groupby('Customer ID')['Profit'].sum().sort_values(ascending=False).reset_index()
customer_profit['Cumulative Profit'] = customer_profit['Profit'].cumsum()
customer_profit['Profit Percentage'] = (customer_profit['Cumulative Profit'] / customer_profit['Profit'].sum()) * 100
customer_profit['Customer Percentage'] = (customer_profit.index + 1) / len(customer_profit) * 100

# Find the point where 80% of profit is reached
pareto_80_profit_idx = (customer_profit['Profit Percentage'] >= 80).idxmax()
pareto_customer_percentage = customer_profit.loc[pareto_80_profit_idx, 'Customer Percentage']

print(f"Environ {pareto_customer_percentage:.2f}% des clients contribuent à 80% du profit total.")

plt.figure(figsize=(12, 6))
plt.plot(customer_profit['Customer Percentage'], customer_profit['Profit Percentage'], marker='o', linestyle='-')
plt.axvline(x=pareto_customer_percentage, color='r', linestyle='--', label=f'{pareto_customer_percentage:.2f}% Clients')
plt.axhline(y=80, color='g', linestyle='--', label='80% Profit')
plt.title('Courbe cumulative du profit par client (Principe de Pareto)', fontsize=16)
plt.xlabel('% de clients', fontsize=12)
plt.ylabel('% de profit cumulé', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

#### Quelles sont les 20 villes les plus populaires selon les ventes ?

```python
sales_by_city = df.groupby('City')['Sales'].sum().sort_values(ascending=False).reset_index()

print("Top 20 villes par ventes :")
display(sales_by_city.head(20))

plt.figure(figsize=(14, 8))
sns.barplot(x='Sales', y='City', data=sales_by_city.head(20), palette='viridis')
plt.title('Top 20 villes par ventes', fontsize=16)
plt.xlabel('Ventes totales', fontsize=12)
plt.ylabel('Ville', fontsize=12)
plt.tight_layout()
plt.show()
```

#### Qu’en est-il des 20 meilleures villes par profit ?

```python
profit_by_city = df.groupby('City')['Profit'].sum().sort_values(ascending=False).reset_index()

print("Top 20 villes par profit :")
display(profit_by_city.head(20))

plt.figure(figsize=(14, 8))
sns.barplot(x='Profit', y='City', data=profit_by_city.head(20), palette='plasma')
plt.title('Top 20 villes par profit', fontsize=16)
plt.xlabel('Profit total', fontsize=12)
plt.ylabel('Ville', fontsize=12)
plt.tight_layout()
plt.show()
```

#### Y a-t-il des différences de rentabilité entre les villes ?

```python
merged_city_data = pd.merge(sales_by_city, profit_by_city, on='City', suffixes=('_Sales', '_Profit'))
merged_city_data['Profit_Margin'] = (merged_city_data['Profit'] / merged_city_data['Sales']) * 100
merged_city_data = merged_city_data.sort_values(by='Sales', ascending=False).reset_index(drop=True)

print("Analyse des ventes, profits et marges bénéficiaires pour les villes principales (trié par ventes) :")
display(merged_city_data.head(20))

# Visualize cities with high sales but potentially low profit margins
plt.figure(figsize=(16, 8))
sns.scatterplot(x='Sales', y='Profit', hue='Profit_Margin', size='Sales', sizes=(50, 1000), data=merged_city_data.head(50), palette='coolwarm')

for i, row in merged_city_data.head(20).iterrows():
    plt.text(row['Sales'], row['Profit'], row['City'], fontsize=9, ha='right')

plt.title('Ventes vs Profit par Ville avec Marge Bénéficiaire (Top 50 Villes)', fontsize=16)
plt.xlabel('Ventes totales', fontsize=12)
plt.ylabel('Profit total', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Marge Bénéficiaire (%)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Analyse des clients

### Quels sont les 20 principaux clients par Ventes ?

```python
customer_sales = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).reset_index()

print("Top 20 clients par ventes :")
display(customer_sales.head(20))

plt.figure(figsize=(14, 8))
sns.barplot(x='Sales', y='Customer Name', data=customer_sales.head(20), palette='crest')
plt.title('Top 20 clients par ventes', fontsize=16)
plt.xlabel('Ventes totales', fontsize=12)
plt.ylabel('Nom du client', fontsize=12)
plt.tight_layout()
plt.show()
```

### Tracez la courbe cumulative des ventes par clients. Peut-on appliquer le principe de Pareto aux clients et aux ventes ?

```python
customer_sales_pareto = df.groupby('Customer ID')['Sales'].sum().sort_values(ascending=False).reset_index()
customer_sales_pareto['Cumulative Sales'] = customer_sales_pareto['Sales'].cumsum()
customer_sales_pareto['Sales Percentage'] = (customer_sales_pareto['Cumulative Sales'] / customer_sales_pareto['Sales'].sum()) * 100
customer_sales_pareto['Customer Percentage'] = (customer_sales_pareto.index + 1) / len(customer_sales_pareto) * 100

# Find the point where 80% of sales is reached
pareto_80_sales_idx = (customer_sales_pareto['Sales Percentage'] >= 80).idxmax()
pareto_customer_percentage_sales = customer_sales_pareto.loc[pareto_80_sales_idx, 'Customer Percentage']

print(f"Environ {pareto_customer_percentage_sales:.2f}% des clients contribuent à 80% des ventes totales.")

plt.figure(figsize=(12, 6))
plt.plot(customer_sales_pareto['Customer Percentage'], customer_sales_pareto['Sales Percentage'], marker='o', linestyle='-')
plt.axvline(x=pareto_customer_percentage_sales, color='r', linestyle='--', label=f'{pareto_customer_percentage_sales:.2f}% Clients')
plt.axhline(y=80, color='g', linestyle='--', label='80% Ventes')
plt.title('Courbe cumulative des ventes par client (Principe de Pareto)', fontsize=16)
plt.xlabel('% de clients', fontsize=12)
plt.ylabel('% de ventes cumulées', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

## Décisions sur les États et Villes à Prioriser pour les Stratégies Marketing

Basé sur l'analyse approfondie des ventes et des bénéfices par État et par Ville, ainsi que l'application du principe de Pareto aux clients, voici les décisions et recommandations pour la stratégie marketing :

### États à Prioriser :

1.  **Priorité Élevée (Excellentes Ventes et Bénéfices)**:
    *   **Californie et New York** : Ces deux États sont les leaders incontestés en termes de ventes et de bénéfices. Ils devraient continuer à être les marchés primaires. Les stratégies marketing devraient viser à maintenir la fidélité des clients existants et à introduire de nouveaux produits ou services. Il est crucial de capitaliser sur les clients exceptionnels identifiés à New York.
    *   **Washington, Michigan, Virginie, Indiana, Géorgie, Kentucky, Minnesota, Delaware** : Ces États affichent des profits élevés et méritent une attention soutenue. Des campagnes de marketing ciblées pourraient aider à consolider leur position et à stimuler davantage la croissance.

2.  **Zones de Surveillance et d'Optimisation (Fortes Ventes, Faibles Bénéfices/Pertes)**:
    *   **Texas, Ohio, Pennsylvanie, Illinois** : Ces États ont montré des volumes de ventes importants mais souffrent de faibles profits, voire de pertes. Cela suggère des problèmes de marges, potentiellement dus à des remises excessives, des coûts logistiques élevés ou une concurrence féroce. La stratégie marketing ici devrait être double :
        *   **Analyse des remises**: Réévaluer les politiques de remise et identifier les points de prix optimaux pour améliorer la rentabilité.
        *   **Optimisation des opérations**: Collaborer avec les équipes d'opérations pour réduire les coûts.
        *   **Marketing ciblé**: Mettre l'accent sur les produits à forte marge et les segments de clientèle les plus rentables dans ces États.

### Villes à Prioriser :

1.  **Villes Performantes (Fortes Ventes et Profits)**:
    *   **New York City, Los Angeles, Seattle, San Francisco, Detroit, Lafayette, Jackson, Atlanta, Minneapolis, San Diego, Springfield, Columbus, Newark, Columbia, Richmond, Providence, Henderson, Arlington, Lakewood, Milwaukee** : Ces villes représentent le cœur de l'activité. Les efforts marketing doivent y être renforcés pour maintenir la part de marché, encourager les achats répétés et attirer de nouveaux clients.

2.  **Villes à Potentiel et à Risque (Divergence Ventes/Profits)**:
    *   **Philadelphie, Houston, Chicago, San Antonio, Lancaster** : Ces villes apparaissent dans le top 20 des ventes mais figurent aussi parmi les villes les moins rentables (ou même en perte). Ceci est une alerte majeure. La stratégie marketing doit être réajustée pour :
        *   **Revoir les prix et les promotions**: Éviter les promotions qui érodent les marges.
        *   **Identifier les segments rentables**: Cibler les clients ou les produits qui génèrent des profits positifs.
        *   **Optimisation des coûts**: Examiner les coûts de distribution et de service client spécifiques à ces villes.

### Principe de Pareto (Clients et Profit/Ventes) :

*   L'analyse a montré qu'environ **19.29% des clients contribuent à 80% du profit total** et un pourcentage similaire de clients contribue à 80% des ventes. Ceci confirme l'applicabilité du principe de Pareto.
*   **Stratégie Marketing Client**:
    *   **Concentration sur les clients clés**: Identifier et fidéliser les **20% de clients** qui génèrent la majorité des profits et des ventes. Des programmes de fidélisation premium, un service client personnalisé et des offres exclusives peuvent être mis en place pour ces clients.
    *   **Campagnes de réactivation**: Pour les clients moins actifs mais ayant un historique d'achat significatif, des campagnes de réengagement peuvent être envisagées.
    *   **Acquisition ciblée**: Utiliser les profils des clients à forte valeur pour cibler l'acquisition de nouveaux clients similaires.

En résumé, la stratégie marketing devrait adopter une approche différenciée : capitaliser sur les marchés et clients performants, mais aussi investiguer et corriger les problèmes de rentabilité dans les zones à fort volume de ventes mais à faible profit.
