```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pl

df = pd.read_csv('/content/sample_data/Student Mental health.csv' )
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a histogram for the 'What is your CGPA?' column
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='What is your CGPA?', color='skyblue')
plt.title('Distribution du CGPA des étudiants')
plt.xlabel('Catégorie de CGPA')
plt.ylabel('Nombre d\'étudiants')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
