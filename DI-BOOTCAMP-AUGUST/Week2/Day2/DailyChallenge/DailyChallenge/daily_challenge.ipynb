```python
import pandas as pd

data = pd.read_csv('/content/sample_data/datascience_salaries.csv',index_col=0)
data.dropna()
print(set(data["salary_currency"]))
data
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data["salary_normalized"]= scaler.fit_transform(data[['salary']])
print(data[["salary_currency", "salary_normalized"]])
```

```python
from sklearn.decomposition import PCA

numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns.tolist()
pca = PCA(n_components=2)
components = pca.fit_transform(data[numeric_cols])
data['PCA1'] = components[:, 0]
data['PCA2'] = components[:, 1]
```

```python
data
```

```python
resultat = data.groupby('experience_level')['salary'].agg(['mean', 'median'])
print(resultat)
```
