EXERCISE 3

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

ds = norm.rvs(loc=50, scale=10, size = 200)

# Calculate mean and standard deviation from the generated sample
mu, std = np.mean(ds), np.std(ds)

# Create the histogram
plt.hist(ds, bins=10, density=True, alpha=0.6, color='g', label='Histogram of Data')

# Plot the PDF. Generate x values for the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2, label='Normal PDF')

title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.stats import f_oneway

fertilizer_1= [5, 6, 7, 6, 5]
fertilizer_2= [7, 8, 7, 9, 8]
fertilizer_3= [4, 5, 4, 3, 4]

f_statistic, p_value = f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Interprétation de la valeur p
alpha = 0.05 # Niveau de signification courant

if p_value < alpha:
    print(f"\nComme la valeur p ({p_value:.5f}) est inférieure à notre niveau de signification alpha ({alpha}),")
    print("nous rejetons l'hypothèse nulle. Cela suggère qu'il existe une différence significative entre les moyennes d'au moins deux groupes d'engrais.")
else:
    print(f"\nComme la valeur p ({p_value:.5f}) est supérieure à notre niveau de signification alpha ({alpha}),")
    print("nous ne rejetons pas l'hypothèse nulle. Cela suggère qu'il n'y a pas de différence significative entre les moyennes des groupes d'engrais.")
```

EXERCISE 8

```python
import pandas as pd
from scipy.stats import pearsonr,spearmanr

data = pd.DataFrame(
      {
          'age': [23, 25, 30, 35, 40],
      'income': [35000, 40000, 50000, 60000, 70000]
      }
    )

corr_pearson, _ = pearsonr(data['age'], data['income'])

print("coefficient pearson:", corr_pearson)

corr_spearman, _ = spearmanr(data['age'], data['income'])

print("coefficient spearman:", corr_spearman)
```

```python
from scipy.stats import binom
import numpy as np

# --- Exemple : probabilité d'obtenir exactement 5 faces en 10 lancers de pièce ---

# Paramètres de la distribution binomiale
n = 10  # Nombre de lancers de pièce (essais)
p = 0.5 # Probabilité d'obtenir une face (succès) pour chaque lancer
k = 5   # Nombre de faces que nous voulons obtenir (nombre de succès)

# Calculer la probabilité d'obtenir exactement k succès (fonction de masse de probabilité - PMF)
probability_exactly_k = binom.pmf(k, n, p)

print(f"Probabilité d'obtenir exactement {k} faces en {n} lancers: {probability_exactly_k:.4f}\n")

```
