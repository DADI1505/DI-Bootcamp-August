```python
import numpy as np

# Création d'une matrice 3x3
matrice = np.array([[6, 1, 1],
                    [4, -2, 5],
                    [2, 8, 7]])

print("Matrice :\n", matrice)

# 1. Calculer le déterminant
determinant = np.linalg.det(matrice)
print(f"\nDéterminant : {determinant:.2f}")

# 2. Trouver l'inverse de la matrice
# On vérifie d'abord si le déterminant n'est pas nul
if determinant != 0:
    inverse = np.linalg.inv(matrice)
    print("\nInverse de la matrice :\n", inverse)
else:
    print("\nLa matrice n'est pas inversible (le déterminant est nul).")
```

```python
import numpy as np

# Générer un tableau de 50 nombres aléatoires (entre 0 et 100 par exemple)
np.random.seed(42)  # Pour la reproductibilité
donnees = np.random.randint(0, 101, 50)

print("Tableau de données :\n", donnees)

# 1. Calculer la moyenne et la médiane
moyenne = np.mean(donnees)
mediane = np.median(donnees)

# 2. Calculer l'écart-type
ecart_type = np.std(donnees)

print(f"\nMoyenne : {moyenne:.2f}")
print(f"Médiane : {mediane:.2f}")
print(f"Écart-type : {ecart_type:.2f}")
```

```python
import pandas as pd
import numpy as np

# 1. Création d'un DataFrame avec des nombres aléatoires
np.random.seed(0)
data_random = np.random.randint(1, 100, size=(10, 3))
df_exercice = pd.DataFrame(data_random, columns=['A', 'B', 'C'])
print("DataFrame original :\n", df_exercice)

# 2. Sélection conditionnelle (ex: lignes où la colonne A > 50)
selection = df_exercice[df_exercice['A'] > 50]
print("\nLignes où la colonne A est supérieure à 50 :\n", selection)

# 3. Agrégation (Somme et Moyenne)
somme_colonnes = df_exercice.sum()
moyenne_colonnes = df_exercice.mean()

print("\nSomme par colonne :\n", somme_colonnes)
print("\nMoyenne par colonne :\n", moyenne_colonnes)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- EXPLICATION ---
# Dans NumPy, une image est représentée par un tableau (ndarray).
# 1. Image en niveaux de gris : Un tableau 2D de forme (hauteur, largeur).
# 2. Image en couleur (RVB) : Un tableau 3D de forme (hauteur, largeur, 3).
# Chaque nombre dans le tableau correspond à l'intensité d'un pixel,
# généralement entre 0 (noir) et 255 (blanc) pour le format uint8.

# Création d'une image 5x5 avec des valeurs aléatoires
np.random.seed(42)
image_gris = np.random.randint(0, 256, (5, 5), dtype=np.uint8)

print("Matrice de l'image (intensités des pixels) :\n", image_gris)

# Visualisation
plt.imshow(image_gris, cmap='gray')
plt.title("Image 5x5 en niveaux de gris")
plt.colorbar()
plt.show()
```

EXERCISE 6

```python
import numpy as np

# 1. Préparation des données
np.random.seed(42)
productivity_before = np.random.normal(loc=50, scale=10, size=30)
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# --- FORMULATION DE L'HYPOTHÈSE ---
# Hypothèse : Le programme de formation a un impact positif,
# augmentant la productivité moyenne d'au moins 5 unités.

# 2. Analyse statistique avec NumPy
moyenne_avant = np.mean(productivity_before)
moyenne_apres = np.mean(productivity_after)
amelioration_moyenne = moyenne_apres - moyenne_avant

# 3. Test de l'efficacité
print(f"Productivité moyenne avant : {moyenne_avant:.2f}")
print(f"Productivité moyenne après : {moyenne_apres:.2f}")
print(f"Amélioration moyenne : {amelioration_moyenne:.2f}")

if amelioration_moyenne > 0:
    print("\nRésultat : L'hypothèse est soutenue. La formation a augmenté la productivité.")
    # Calcul du pourcentage d'employés ayant progressé
    progressions = productivity_after > productivity_before
    pourcentage_succes = np.mean(progressions) * 100
    print(f"Pourcentage d'employés ayant amélioré leur score : {pourcentage_succes:.1f}%")
else:
    print("\nRésultat : L'hypothèse n'est pas soutenue. Aucun gain de productivité détecté.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Création de données avec NumPy
# On génère 20 points de données aléatoires
np.random.seed(42)
x = np.arange(20)
y = np.random.randint(1, 50, size=20)

# 2. Visualisation avec Matplotlib
plt.figure(figsize=(10, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Valeurs aléatoires')

# Personnalisation du graphique
plt.title("Graphique linéaire de nombres aléatoires (NumPy + Matplotlib)")
plt.xlabel("Index")
plt.ylabel("Valeur")
plt.grid(True, linestyle='--')
plt.legend()

# Affichage
plt.show()
```
