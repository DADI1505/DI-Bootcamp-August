```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Data Preparation
# Generate synthetic data: 10 cities (rows) x 12 months (columns)
np.random.seed(42) # For reproducibility
data = np.random.uniform(-5, 35, (10, 12))

cities = [f'City_{i+1}' for i in range(10)]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

df = pd.DataFrame(data, index=cities, columns=months)
display(df)
```

```python
# 2. Data Analysis
# Calculate annual mean temperature for each city
df['Annual_Mean'] = df.mean(axis=1)

# Identify hottest and coldest cities
hottest_city = df['Annual_Mean'].idxmax()
coldest_city = df['Annual_Mean'].idxmin()

print(f"Average annual temperatures:\n{df['Annual_Mean']}")
print(f"\nHottest City: {hottest_city} ({df.loc[hottest_city, 'Annual_Mean']:.2f}°C)")
print(f"Coldest City: {coldest_city} ({df.loc[coldest_city, 'Annual_Mean']:.2f}°C)")


```

```python
import matplotlib.pyplot as plt

# 3. Visualisation des données : Graphique linéaire
plt.figure(figsize=(14, 7))

# On itère sur chaque ville pour tracer sa ligne
for city in cities:
    plt.plot(months, df.loc[city, months], marker='o', label=city)

plt.title('Tendances des températures mensuelles par ville')
plt.xlabel('Mois')
plt.ylabel('Température (°C)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
