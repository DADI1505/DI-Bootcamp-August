##  Exercice 1 : Réécrire et optimiser une invite vague

**Scénario :** Le manager demande un post LinkedIn pour promouvoir l'application FlowNest, avec la consigne vague :
> « Écris quelque chose sur les conseils de productivité. »

### Tâches
- Identifier 3 problèmes critiques avec cette consigne.
- La réécrire en utilisant : l'incitation de rôle, l'incitation basée sur des instructions claires, et une contrainte de format/ton/longueur.
- Ajouter une dernière phrase qui rend la consigne réutilisable.

```python
# On liste les 3 problèmes critiques de la consigne vague
# (indice donné : voix de la marque, utilisateurs cibles, engagement sur la plateforme)
problemes_critiques = [
    "Aucune mention du produit ou de la marque : le nom 'FlowNest' n'apparaît nulle part, donc le contenu généré sera générique et non personnalisé.",
    "Public cible non défini : on ne sait pas si le post s'adresse à des étudiants, des professionnels ou des managers, ce qui empêche d'adapter le ton et le vocabulaire.",
    "Aucune contrainte de format, de ton ou de longueur : rien n'indique le style attendu (professionnel, inspirant...), la longueur du post, ni un éventuel appel à l'action, ce qui rend le résultat imprévisible pour un post LinkedIn."
]

# On affiche chaque problème avec un numéro
for i, probleme in enumerate(problemes_critiques, start=1):
    print(f"{i}. {probleme}")
```

```python
# On réécrit la consigne en combinant 3 techniques :
# 1) incitation de rôle ("Agis en tant que...")
# 2) instructions claires ("Donne 3 conseils concrets...")
# 3) contrainte de format/ton/longueur (puces, ton, nombre de mots)
# + une dernière phrase qui rend la consigne réutilisable

consigne_amelioree = """Agis en tant qu'expert en marketing sur les réseaux sociaux pour une startup technologique.
Rédige un post LinkedIn engageant pour promouvoir notre nouvelle application de productivité, FlowNest.

Donne 3 conseils concrets et applicables qui montrent comment FlowNest aide les utilisateurs à mieux se concentrer et à gérer leur temps.

Utilise un ton amical, inspirant et professionnel, adapté à un public LinkedIn de professionnels actifs.
Présente les conseils sous forme de puces, termine par un appel à l'action clair, et reste sous 150 mots. Ajoute 2 à 3 hashtags pertinents.

Utilise ce même format chaque fois que nous faisons la promotion d'outils de productivité sur les réseaux sociaux."""

print(consigne_amelioree)
```

##  Exercice 2 : Prompt en plusieurs parties pour la génération de quiz

**Scénario :** Un professeur de sciences en 5e veut un contenu éducatif (résumé + quiz) à partir d'un article sur les éruptions volcaniques.

### Tâches
- Rédiger un prompt qui demande : un résumé en 2 points, 3 QCM (1 bonne réponse + 2 distracteurs), un ton amical adapté aux 11-13 ans, et une sortie prête pour Google Slides.
- Comparer avec la consigne vague « Faites un quiz pour les enfants sur cet article. »
- Expliquer 3 façons dont la nouvelle version évite un résultat de mauvaise qualité.

```python
# Prompt complet en plusieurs parties pour générer le contenu éducatif
consigne_quiz = """Tu es un professeur de sciences sympathique qui s'adresse à des élèves de 5e (11 à 13 ans).
À partir de l'article fourni sur les éruptions volcaniques, effectue les tâches suivantes :

1. Résume l'article en exactement 2 points simples, avec un vocabulaire facile à comprendre.
2. Crée 3 questions à choix multiples sur l'article. Chaque question doit avoir 1 bonne réponse et 2 réponses fausses mais plausibles (distracteurs). Indique clairement la bonne réponse pour chaque question.
3. Utilise un ton amical et un vocabulaire simple, adapté à des enfants de 11 à 13 ans.
4. Présente le résultat dans un format prêt à être copié-collé dans Google Slides (titres clairs, puces).

Article : [Insérer le texte de l'article ici]"""

print(consigne_quiz)
```

```python
# Trois raisons pour lesquelles cette consigne évite un résultat inutile ou de faible qualité,
# comparée à la consigne vague "Faites un quiz pour les enfants sur cet article."
raisons_amelioration = [
    "Elle précise le TYPE et la QUANTITÉ de contenu attendu (2 points de résumé, exactement 3 QCM avec distracteurs), ce qui évite un résultat trop générique ou incomplet.",
    "Elle définit précisément l'âge et le ton du public visé (11-13 ans, ton amical), contrairement à la consigne vague 'pour les enfants' qui ne précise ni l'âge ni le registre de langage.",
    "Elle impose un format de sortie clair et exploitable (prêt pour Google Slides, avec titres et puces), ce qui garantit un contenu directement utilisable plutôt qu'un texte brut mal structuré."
]

for i, raison in enumerate(raisons_amelioration, start=1):
    print(f"{i}. {raison}")
```

## Exercice 3 : Ajoutez du contexte, obtenez de meilleurs résultats

**Scénario :** Un stagiaire donne la consigne vague « Summarize this report. » pour préparer un résumé financier de 3 minutes, et obtient un résumé générique sans chiffres utiles.

### Tâches
1. Remplir un tableau identifiant le contexte manquant et ce qu'il faut ajouter.
2. Réécrire la consigne du stagiaire en incluant au moins 4 types de contexte.

```python
# Tableau identifiant, pour chaque type de contexte, s'il est manquant et ce qu'il faut ajouter
tableau_contexte = {
    "Role": {"manquant": True, "a_ajouter": "Agis en tant qu'analyste financier."},
    "Audience": {"manquant": True, "a_ajouter": "Ce résumé s'adresse à une équipe de direction non technique."},
    "Purpose": {"manquant": True, "a_ajouter": "Préparer un résumé oral de 3 minutes pour la réunion mensuelle de mise à jour financière."},
    "Input Source": {"manquant": True, "a_ajouter": "Il s'agit du rapport financier trimestriel du T1 2023."},
    "Format/Style": {"manquant": True, "a_ajouter": "Présenter 3 points clés, chacun appuyé par des données chiffrées, dans un langage simple."},
    "Constraints": {"manquant": True, "a_ajouter": "Se limiter à 3 points clés, éviter le jargon technique, rester concis pour une présentation de 3 minutes."},
}

for type_contexte, details in tableau_contexte.items():
    print(f"{type_contexte} -> manquant: {details['manquant']} | à ajouter: {details['a_ajouter']}")
```

```python
# Réécriture de la consigne du stagiaire en incluant au moins 4 types de contexte
# (rôle, audience, objectif, source des données, format/style, contraintes)
consigne_finance = """Agis en tant qu'analyste financier. Ta tâche est de résumer le rapport financier trimestriel du T1 2023 fourni ci-dessous.

Ce résumé s'adresse à une équipe de direction non technique et sera utilisé pour une présentation orale de 3 minutes lors de la réunion mensuelle de mise à jour financière.

Présente exactement 3 points clés, chacun appuyé par des données chiffrées pertinentes, dans un langage simple et sans jargon technique. Reste concis pour respecter le format de 3 minutes.

Rapport financier : [Insérer le texte du rapport ici]"""

print(consigne_finance)
```

##  Exercice 4 : Adapter le prompt à son objectif (Match Prompt to Purpose)

**Scénario :** Construire un template de prompt pour un chatbot de support client dans une entreprise de e-commerce, en choisissant un style parmi : Exploratoire, Structuré, Conversationnel, Fonctionnel.

### Tâches
- Écrire le prompt complet dans le style choisi.
- Identifier 2 caractéristiques clés qui correspondent au style.
- Justifier pourquoi ce style est meilleur que les autres pour ce cas précis.

```python
# On choisit le style "Fonctionnel", car un chatbot de support client doit avant tout
# résoudre des problèmes précis plutôt que discuter ou brainstormer

style_choisi = "Fonctionnel"

modele_de_consigne = """Tu es un chatbot de support client pour une entreprise de commerce électronique spécialisée en électronique. Ta mission principale est de résoudre les problèmes des clients de façon efficace et précise.

Quand un client décrit un problème, identifie le nom du produit, le problème principal, et propose une solution précise, étape par étape, à partir de notre base de connaissances. Si aucune solution n'est trouvée, transfère la demande à un agent humain en créant un ticket résumant le produit et le problème.

Ta réponse doit être directe, informative, et se terminer par une proposition d'aide supplémentaire. N'engage pas de conversation informelle et ne donne pas d'informations non vérifiées."""

# Les 2 caractéristiques clés qui montrent que ce prompt correspond bien au style Fonctionnel
caracteristiques_cles = [
    "Il exécute des tâches précises (identifier le produit/problème, proposer une solution, transférer si besoin) plutôt que de simplement discuter.",
    "Il privilégie la précision et la fiabilité de la réponse plutôt que le ton conversationnel ou la génération d'idées."
]

# Justification du choix de ce style plutôt que les autres
justification = """Le style Fonctionnel est le plus adapté à un chatbot de support client car son objectif principal est d'exécuter des tâches précises (dépannage, proposition de solutions, transfert vers un humain) avec fiabilité. Contrairement aux styles Exploratoire (remue-méninges), Structuré (format fixe) ou Conversationnel (ton humain et bavard), le style Fonctionnel se concentre sur la résolution directe des problèmes, ce qui est essentiel pour un service client efficace et évite toute ambiguïté."""

print(style_choisi)
print(modele_de_consigne)
print(caracteristiques_cles)
print(justification)
```

##  Exercice 5 : Défi de raffinement du prompt — Contrôler le style, la structure et la longueur

**Scénario :** Rédiger un texte publicitaire pour la montre connectée PulseOne Mini, avec un ton amical, un format en puces, un maximum de 50 mots, et 3 caractéristiques obligatoires (batterie, suivi d'activité, Bluetooth).

### Tâches
1. Écrire un prompt qui impose strictement ces contraintes.
2. Évaluer : le nombre de mots est-il respecté ? Les 3 caractéristiques sont-elles mentionnées ? Le ton est-il amical ?
3. Si besoin, réécrire le prompt avec un langage de contrainte plus fort.

 Dans la version originale de ce notebook, la cellule `eval_v2` avait été collée par erreur comme cellule **markdown** au lieu de cellule **code** : elle ne s'exécutait donc jamais. C'est corrigé ci-dessous.

```python
# Première version du prompt : les contraintes sont présentes mais formulées de façon "douce"
consigne_pulseone_v1 = """Rédige un texte publicitaire pour la nouvelle montre connectée 'PulseOne Mini'. Le ton doit être amical. Présente les informations sous forme de puces. Le texte ne doit pas dépasser 50 mots au total. Assure-toi de mentionner sa très bonne autonomie de batterie, ses fonctionnalités avancées de suivi d'activité physique, et sa compatibilité Bluetooth."""

print(consigne_pulseone_v1)
```

```python
# Évaluation de la consigne v1 (à confirmer après une vraie exécution par un LLM)
evaluation_v1 = {
    "nombre_de_mots_respecte": False,  # les LLM respectent rarement une limite de mots stricte si elle n'est pas formulée avec insistance
    "fonctionnalites_mentionnees": True,  # les 3 caractéristiques sont bien listées dans la consigne
    "ton_amical_respecte": True,  # "le ton doit être amical" est une instruction claire
    "remarques": "La consigne précise clairement les fonctionnalités et le ton attendus. Cependant, imposer une limite stricte de 50 mots reste difficile pour un LLM avec une formulation douce ('ne doit pas dépasser'), et nécessite souvent une contrainte plus insistante."
}

evaluation_v1
```

```python
# Deuxième version : on renforce le langage de contrainte pour forcer le respect de la limite de mots
consigne_pulseone_v2 = """Rédige un texte publicitaire pour la nouvelle montre connectée 'PulseOne Mini'. Le ton doit être amical et engageant. Produis STRICTEMENT un maximum de 50 mots, sous forme de puces. Mentionne obligatoirement : l'autonomie de la batterie, le suivi d'activité physique, et la compatibilité Bluetooth. NE DÉPASSE PAS 50 mots, quelles que soient les circonstances."""

print(consigne_pulseone_v2)
```

```python
# Évaluation de la consigne v2, avec une contrainte de longueur beaucoup plus stricte
evaluation_v2 = {
    "nombre_de_mots_respecte": "À vérifier après exécution par un LLM",  # on s'attend à un meilleur respect grâce à la formulation plus stricte
    "fonctionnalites_mentionnees": True,
    "ton_amical_respecte": True,
    "remarques": "L'ajout de 'STRICTEMENT' et 'NE DÉPASSE PAS... quelles que soient les circonstances' rend la contrainte de longueur beaucoup plus insistante, ce qui devrait améliorer le respect de la limite de 50 mots. Il faudrait néanmoins exécuter cette consigne avec un vrai LLM pour confirmer."
}

evaluation_v2
```

##  Exercice 6 : Détecter et atténuer les hallucinations

**Scénario :** En résumant un article scientifique sur le changement climatique et la biodiversité marine, ChatGPT ajoute une affirmation qui n'apparaît nulle part dans l'article original.

### Tâches
1. Écrire un prompt précis demandant au modèle d'éviter les hallucinations et de n'utiliser que le texte fourni.
2. Réécrire le prompt en utilisant un langage de vérification.
3. Lister 2 stratégies d'atténuation utilisées.
4. Nommer 2 domaines professionnels où les hallucinations sont particulièrement risquées, et expliquer pourquoi.

 Dans la version originale, cet exercice n'était pas terminé (des TODO vides `'[WRITE PROMPT...]'`). Il est maintenant complété ci-dessous.

```python
# Prompt de base demandant au modèle de rester "ancré" dans le texte fourni,
# sans ajouter d'information qui n'y figure pas
consigne_anti_hallucination = """Résume l'article scientifique ci-dessous sur le changement climatique et la biodiversité marine.

Utilise UNIQUEMENT les informations explicitement présentes dans le texte fourni. N'ajoute aucune information, statistique ou affirmation qui ne figure pas littéralement dans l'article, même si elle te semble plausible.

Article : [Insérer le texte de l'article ici]"""

print(consigne_anti_hallucination)
```

```python
# Version réécrite avec un langage de vérification explicite,
# qui demande au modèle de signaler ce qu'il ne sait pas plutôt que d'inventer une réponse
consigne_verification = """Résume l'article scientifique ci-dessous sur le changement climatique et la biodiversité marine.

N'inclus que des affirmations explicitement énoncées dans l'article. Si une information n'est pas clairement présente dans le texte, indique "information non disponible dans l'article" plutôt que de l'inventer.

Avant de conclure, relis chaque phrase de ton résumé et vérifie qu'elle peut être directement retrouvée dans le texte source.

Article : [Insérer le texte de l'article ici]"""

print(consigne_verification)
```

```python
# Deux stratégies d'atténuation utilisées dans les prompts ci-dessus
strategies_attenuation = [
    "Demander explicitement au modèle de se limiter aux informations présentes dans le texte fourni, sans en ajouter.",
    "Demander au modèle d'indiquer clairement quand une information n'est pas disponible, plutôt que de deviner ou d'inventer une réponse."
]

# Deux domaines professionnels où les hallucinations sont particulièrement risquées, et pourquoi
domaines_a_risque = {
    "domaine_1": {
        "nom": "Juridique",
        "pourquoi": "Une hallucination dans un résumé juridique (par exemple inventer une jurisprudence ou un article de loi) peut conduire à des décisions erronées avec de graves conséquences légales."
    },
    "domaine_2": {
        "nom": "Santé",
        "pourquoi": "Une information médicale inventée (dosage, diagnostic, effet secondaire) peut directement mettre en danger la santé ou la vie d'un patient."
    }
}

print(strategies_attenuation)
print(domaines_a_risque)
```
